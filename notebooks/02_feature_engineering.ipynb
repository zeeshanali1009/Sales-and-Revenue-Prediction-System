```python
# Creating Year-Month Column
import pandas as pd

df = pd.read_csv("../data/raw/superstore.csv", encoding="latin1")
df['Order Date'] = pd.to_datetime(df['Order Date'])

```

```python
# Create YearMonth
df['YearMonth'] = df['Order Date'].dt.to_period('M')
```

```python
# Aggregate Monthly Sales
monthly_df = df.groupby('YearMonth').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Discount': 'mean'
}).reset_index()
```

```python
# Convert Period to Timestamp
monthly_df['YearMonth'] = monthly_df['YearMonth'].dt.to_timestamp()
monthly_df.head()
```

```python
# Feature Engineering
monthly_df['Year'] = monthly_df['YearMonth'].dt.year
monthly_df['Month'] = monthly_df['YearMonth'].dt.month
```

```python
# Lag Features
monthly_df['Sales_Lag_1'] = monthly_df['Sales'].shift(1)
monthly_df['Sales_Lag_2'] = monthly_df['Sales'].shift(2)
monthly_df['Sales_Lag_3'] = monthly_df['Sales'].shift(3)
```

```python
# Rolling Mean (Smoothing the trend)
monthly_df['Sales_Rolling_3'] = monthly_df['Sales'].rolling(3).mean()
```

```python
# Drop NaNs
monthly_df.dropna(inplace=True)
```

```python
# Train-Test Split
train_size = int(len(monthly_df) * 0.8)

train_df = monthly_df.iloc[:train_size]
test_df = monthly_df.iloc[train_size:]
```

```python
# Saved Processed data
monthly_df.to_csv("../data/processed/monthly_sales.csv", index=False)
```
